```python
import pandas as pd

# maptlotlib
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.rc_context(rc={"figure.figsize": (10, 8)})

# seaborn
import seaborn as sns
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (12, 10)})
sns.set_style("darkgrid")

from datasets import get_data
attr, test_attr, target = get_data()
```

## Correlation Between Features
There is a high correlation between many of the "unknown" features. Might want to apply PCA to separate them as much as possible.

```python
attr0 = attr.copy()
attr0['target'] = target
sns.corrplot(attr, annot=False, diag_names=False)
```

## Grid Comparison of Variables (Color = Revenue)

```python
pairs = ['City', 'Year', 'P1', 'target']
g = sns.pairplot(attr0[pairs], hue="target", palette="GnBu_d")
```

## Kernel Density Estimates for Each Variable

```python
sns.set_context(rc={"figure.figsize": (8, 4)})
for col in attr0.columns:
    sns.kdeplot(attr0[col], shade=True, label=col)
    plt.show()
```

## Compare Each Variable to Revenue (Target)

```python
ignore_cols = ['target', 'P7', 'P10', 'P13']

for col in attr0.columns:
    if col not in ignore_cols:
        sns.jointplot(attr0['target'], attr0[col])
        plt.show()
```
